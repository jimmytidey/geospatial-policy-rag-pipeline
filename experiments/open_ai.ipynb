```python
import openai
import os


def read_first_2000_lines(file_path):
    lines = []
    try:
        with open(file_path, "r") as file:
            for _ in range(600):
                line = file.readline()
                if not line:  # End of file reached before 2000 lines
                    break
                lines.append(line.strip())
    except Exception as e:
        print(f"An error occurred: {e}")
        return ""

    return "\n".join(lines)


# Example usage
file_path = "policy.txt"


# Set your OpenAI API key
api_key = os.environ["OPENAI_API_KEY"]

# Initialize the OpenAI API client
openai.api_key = api_key

# Define the message you want to send
message = """
You are an expert at examining policy documents and splitting them into chunks.
Take the following text, break it into semantic chunks of at most 500 words. Start a new chunk when a new subheading starts.
Ignore page breaks. Please ensure lists do not flow between chunks.
Return an JSON array of objects for each chunk.
Each object should have a field called 'text' with the text of the chunk   
For each chunk, please also a field 'headings', which is an array of all the headings and subheadings this chunk falls under. 
please indicate if the chunk inlcudes a specific planning policy, with a boolean field 'policy'.
If the policy seems to be about an explicitly stateed location, or is site specific, mark a boolean field 'site' as true. 

Text to chunk:
"""


message += read_first_2000_lines(file_path)

# Send the message to ChatGPT-4 and get a response
completion = openai.chat.completions.create(
    model="gpt-4o",
    messages=[
        {
            "role": "user",
            "content": message,
        },
    ],
)
print(completion.choices[0].message.content)
```

```python
def count_lines_in_file(file_path):
    with open(file_path, "r") as file:
        line_count = sum(1 for line in file)
    return line_count


file_path = "policy.txt"  # Replace with your file path
line_count = count_lines_in_file(file_path)
print(f"The file has {line_count} lines.")
```
