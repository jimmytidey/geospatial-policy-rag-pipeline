```python
from unidecode import unidecode
import fitz

doc = fitz.open("files/4_after_processing/London_Borough_of_Newham.pdf")


output = []

for page in doc:
    output += page.get_text("blocks")

previous_block_id = 0  # Set a variable to mark the block id
for block in output:

    print("\n ------------------ \n")
    plain_text = unidecode(block[4])
    print(plain_text)
```

```python
def get_bookmarks(filepath: str):
    # WARNING! One page can have multiple bookmarks!
    bookmarks = {}
    with fitz.open(filepath) as doc:
        toc = doc.get_toc()  # [[lvl, title, page, …], …]
        for level, title, page in toc:
            bookmarks[page] = title
    return bookmarks


result = get_bookmarks("files/4_after_processing/London_Borough_of_Newham.pdf")

print(result)
```
