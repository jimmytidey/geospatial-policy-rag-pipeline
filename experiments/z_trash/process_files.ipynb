```python
from helpers import pdf_to_chunks, embed_send_to_db, list_files
import os
import shutil
import pandas as pd

source_path = file_path = os.path.join("files", "2_before_processing")
destination_path = file_path = os.path.join("files", "4_after_processing")
failed_path = os.path.join("files", "3_failed_processing")
after_path = os.path.join("files", "4_after_processing")
meatdata_file_path = os.path.join("files", "metadata.csv")

# set metadata for every file in the source folder (will be overwritten by the CSV if a match is found)
metadata_global = {"type": "localplan"}

# get metadata per file from the CSV

if os.path.exists(meatdata_file_path):
    metadata_csv_exists = True
    metadata_df = pd.read_csv(meatdata_file_path)
else:
    metadata_csv_exists = False

# start processing the pdfs
pdfs = [f for f in os.listdir(source_path) if not f.startswith(".")]

for pdf in pdfs:
    print(f"Processing {pdf}")

    pdf_path = os.path.join(source_path, pdf)
    try:
        chunks = pdf_to_chunks(pdf_path)

        if metadata_csv_exists:
            print("metadata found in CSV")
            pdf_metadata = metadata_df.loc[metadata_df["filename"] == pdf].to_dict(
                orient="records"
            )[0]

            metadata = metadata_global | pdf_metadata

        else:
            print("NO metadata found in CSV")
            metadata = metadata_global

        print(metadata)

        embed_send_to_db(chunks, pdf, metadata)

        print("*** COMPLETE ***")

        shutil.move(os.path.join(source_path, pdf), destination_path)
    except Exception as e:
        shutil.move(os.path.join(source_path, pdf), failed_path)
        print(f"Error processing {pdf}: {e}")
        continue
```
