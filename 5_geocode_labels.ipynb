```python
from data_pipeline.geocode_geo_labels.geocode_extract import  count_chunks_for_geocoding,extract_chunks_for_geocoding
from data_pipeline.geocode_geo_labels.geocode_transform import geocode_chunk
from data_pipeline.geocode_geo_labels.geocode_load import load_locations

count = count_chunks_for_geocoding()
print("Total chunks ungeocoded: " + str(count))

chunks  = extract_chunks_for_geocoding(1)

if chunks is not None:
    for index, chunk in enumerate(chunks):
        
        geocoded_locations = geocode_chunk(chunk)
        load_locations(geocoded_locations)
        
else:
    print("No records found.")
    
```
