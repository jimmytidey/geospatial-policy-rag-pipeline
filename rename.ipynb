```python
import os
from postgres import Postgres

# Initialize the Postgres connection
pg = Postgres()

# Define the directory containing the files
files_directory = "files"

# Step 1: Fetch all rows from the documents table
query = """
SELECT document_id, category, lpa, neighbourhood, file
FROM documents
"""
rows = pg.query(query)

# Step 2: Process each row and rename files
for row in rows:
    # Extract columns
    document_id = row['document_id']
    category = row['category'] or ""
    lpa = row['lpa'] or ""
    neighbourhood = row['neighbourhood'] or ""
    original_file = row['file']


    # Construct the new file name
    file_name_parts = [category, lpa, neighbourhood, str(document_id)]
    new_file_name = "_".join(filter(None, file_name_parts)) + ".pdf"

   # Update the database with the new file name
    update_query = """
    UPDATE documents
    SET file = %s
    WHERE document_id = %s
    """
    
    print(update_query)
    print (new_file_name)
    print (document_id)
    pg.insert(update_query, (new_file_name, document_id))

    # Rename the file on disk
    original_file_path = os.path.join(files_directory, original_file)
    new_file_path = os.path.join(files_directory, new_file_name)
    if os.path.exists(original_file_path):
        
        os.rename(original_file_path, new_file_path)

 

# Step 3: Completion message
print("File renaming and database updates completed.")
```
